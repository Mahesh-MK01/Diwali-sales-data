# Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x= "Gender",data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum()
ab = sns.barplot(x= 'Gender' , y = 'Amount', data = sales_gen )

for bars in ab.containers:
    ab.bar_label(bars)
```

# Age

```python
df.columns
```

```python
sns.countplot (y='Age Group',hue='Gender', data=df) 
```

```python
ax= sns.countplot(data = df, x='Age Group' , hue ='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by="Age Group", ascending =True)
sns.barplot(x= 'Age Group', y= 'Amount' , data= sales_age)
```

# state

```python
df.columns
```

```python
Order_state= df.groupby (['State'],as_index= False)['Orders'].sum().head(11)
sns.set(rc={'figure.figsize': (17,5)})
sns.barplot(x= 'State',y = 'Orders' , data=Order_state)
```

```python
sale_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y= "Amount", data=sale_state)
```
